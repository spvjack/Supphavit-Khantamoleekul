```python
import pandas as pd
url = "https://raw.githubusercontent.com/chiharuishida/Tutorialdata/main/ad_dataset_withheaders.csv"
AD = pd.read_csv(url)
```

```python
AD.head(10)
#provides the first 10 raw of data. this provides a quick view of the dataset titled "AD"
```

```python
AD.describe()
#quickly provides a descriptive statistics for all the continous variables in the dataset
```

```python

AD.info()
#provides information about each variable in the dataset. Dtype of float is a number with a decimal place, while int is for integers
```

```python
AD['Gender'].value_counts()
#"day" is a categorical variable in our data
```

```python
AD.corr()
#quickly provides a correlation matrix
```

```python
sns.heatmap(AD.corr(), annot=True)
#annot=True adds an annotation to the right
```

```python
sns.heatmap(AD.corr(), cmap="YlGnBu")
#just changing colors here for better looks (darker=greater value)
#other colormap options are 'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds','YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu','GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'
```

```python
sns.pairplot(AD)
#quickly provides pairplot between the continous varialbles.
```

```python
sns.scatterplot(x="Gender", y="Site_Time", data=AD)
#let's just look at the scatterplot between "Gender" and "Site_Time"
#A scatterplot is for any two continuous variables
```

```python
sns.lmplot(x = "Age", y = "Site_Time", data = AD)
#fitting a trendline
```

```python
sns.scatterplot(x="Age", y="DailyTime", hue="AdClicking", data=AD)
#adding color-coding by AdClicking-nonAdClicking by adding the hue statement
```

```python
sns.lmplot(x = "Age", y = "DailyTime", hue="AdClicking", data = AD)
#Fitting lines for the two AdClicking categories
```

```python
sns.countplot(x='Age', data=AD)
#barchart sorted by "Age" which is a categorical variable
```

```python
sns.countplot(x='Age', data=AD, color='b')
#changing color to all red by adding "color='r'"
```

```python
AD_Age=AD.groupby("Age")
AD_Age.mean()
#groupby() is used to split the data in dataframe into groups ("AD_Age" is created as a new dataframe)
#"Age" is a categorical variable
```

```python
sns.boxplot(x="Age", y="Gender", data=AD)
#quickly creating boxplots. Here, "Age" is a categorical variable" and "Gender" is a continuous variable
```

```python

sns.boxplot(x="Age", y="Gender", hue="AdClicking", data=AD)
#adding a color code by using hue. In this case, we are using the "AdClicking" variable, which is a categorical variable
```

```python
AD_top3_AD = AD.nlargest(3, "Age")
AD_top3_AD
#.nlargest()is a method to quickly get TopN results for the continous variable of "tip". It will list the top 3 tippers. For this analysis, select a continuous variable
```
